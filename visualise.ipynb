```python
# Load the data from your csv
df = pd.read_csv('D:/Python_Projects/hav_cpp_from_python_2/cmake-build-debug/output.csv')

gps_x = df['longitude']
gps_y = df['latitude']
yaw = df['yaw']
state_yaw = df['state_yaw']

fig, axs = plt.subplots(2, figsize=(10,10))

# Downsample the data for better visualization (if it's too dense)
downsample_rate = 10 # Change this value to get the best visual result
gps_x = gps_x[::downsample_rate]
gps_y = gps_y[::downsample_rate]
yaw = yaw[::downsample_rate]
state_yaw = state_yaw[::downsample_rate]

# Plot the GPS data with yaw
axs[0].plot(gps_x, gps_y, label='GPS', color='red')
axs[0].quiver(gps_x, gps_y, np.cos(yaw), np.sin(yaw), color='b', scale=20)
axs[0].legend()
axs[0].set_title('GPS with Yaw')

# Plot the GPS data with state_yaw
axs[1].plot(gps_x, gps_y, label='GPS', color='red')
axs[1].quiver(gps_x, gps_y, np.cos(state_yaw), np.sin(state_yaw), color='b', scale=20)
axs[1].legend()
axs[1].set_title('GPS with State_Yaw')

plt.tight_layout()
plt.show()
```

```python
# Load the data from your csv
df = pd.read_csv('D:/Python_Projects/hav_cpp_from_python_2/cmake-build-debug/output.csv')

gps_x = df['longitude']
gps_y = df['latitude']
yaw = df['yaw']
state_yaw = df['state_yaw']

fig, ax = plt.subplots(figsize=(10,10))

# Downsample the data for better visualization (if it's too dense)
downsample_rate = 10 # Change this value to get the best visual result
gps_x = gps_x[::downsample_rate]
gps_y = gps_y[::downsample_rate]
yaw = yaw[::downsample_rate]
state_yaw = state_yaw[::downsample_rate]

# Plot the GPS data with yaw and state_yaw
ax.plot(gps_x, gps_y, label='GPS', color='black')
ax.quiver(gps_x, gps_y, np.cos(yaw), np.sin(yaw), color='b', scale=20, label='Yaw')
ax.quiver(gps_x, gps_y, np.cos(state_yaw), np.sin(state_yaw), color='r', scale=20, label='State_Yaw')
ax.legend()
ax.set_title('GPS with Yaw and State_Yaw')

plt.tight_layout()
plt.show()
```
